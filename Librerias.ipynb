## REGRACIÓN LINEAL



```python
# Grafica de dispersión, : Salarios vs experiencia
#la primera parte es del eje X u la segunda del eje Y

Y = salarios["Salario"]
X = salarios ["AñosdeExperiencia"]
#scatter  es un tipo de grafico de dispersión
## el primer valor es el eje X el segundo del eje Y para poder generar la grafica

plt.scatter(X,Y,color = "red")

#titulo del grafico

plt.title("Salario vs AñosdeExperiencia")

#Titulo Eje X

plt.xlabel("Años de Experiencia")

#Titulo eje Y

plt.ylabel("Salarios")

plt.show()

```
